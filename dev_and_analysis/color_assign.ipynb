```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image_pt = "../output_videos/cropped_img.jpg"
image = cv2.imread(image_pt)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
top_half_img = image[0:int(image.shape[0]/2), :]
plt.imshow(top_half_img)
plt.show()
```

```python
img_2d = image.reshape(-1,3)

kmeans = KMeans(n_clusters=2,random_state=2)
kmeans.fit(img_2d)

labels= kmeans.labels_

clustered_img = labels.reshape(image.shape[0],image.shape[1])

plt.imshow(clustered_img)
plt.show()
```

```python
corner_clusters = [clustered_img[0,0],clustered_img[0,-1],clustered_img[-1,0],clustered_img[-1,-1]]
non_player_cluster = max(set(corner_clusters),key=corner_clusters.count)
print(non_player_cluster)
```

```python
player_cluster = 1-non_player_cluster
print(player_cluster)
```

```python
kmeans.cluster_centers_[player_cluster]
```
